```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
fname = 'smallexp_n3_steps20_eps300_sassy0.csv'
df = pd.read_csv('logs/{}'.format(fname))
```

```python
# Plot the responses for different events and regions
sns.set(font_scale=1.5)
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(12, 4)
sns.lineplot(x="ep", y="R", data=df, ax=axes[0])
axes[0].set(xlabel='Episode', ylabel='Episode Return');
sns.lineplot(x="ep", y="P", data=df, ax=axes[1])
axes[1].set(xlabel='Episode', ylabel='Episode Performance');
plt.tight_layout()
```
