```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
fname = '20181016_n25_steps100_eps300_gamma0.95_sassy0.csv'
df = pd.read_csv('logs/{}'.format(fname))
```

```python
# Plot the responses for different events and regions
sns.set(font_scale=1.5)
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(12, 4)
sns.lineplot(x="step", y="R", data=df, ax=axes[0])
axes[0].set(xlabel='timestep', ylabel='episode return');
axes[0].ticklabel_format(style='sci', scilimits=(0,0), axis='x')
sns.lineplot(x="step", y="P", data=df, ax=axes[1])
axes[1].set(xlabel='timestep', ylabel='episode performance');
plt.tight_layout()
axes[1].ticklabel_format(style='sci', scilimits=(0,0), axis='x')

fig.suptitle(fname) # or plt.suptitle('Main title')
```
